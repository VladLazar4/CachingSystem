```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['1K users', '10K users', '100K users']
postgres = [50.278, 104.67, 733.69]
redis = [6.258, 11.554, 39.412]

x = np.arange(len(labels))  # [0, 1, 2]
width = 0.35                # width of each bar

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, postgres, width, label='PostgreSQL')
bars2 = ax.bar(x + width/2, redis, width, label='Redis')

# Labels and styling
ax.set_xlabel('Dataset size')
ax.set_ylabel('Average response time (ms)')
ax.set_title('Redis vs PostgreSQL Performance')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Optional: display values on top of bars
ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)

plt.tight_layout()
plt.show()
```
